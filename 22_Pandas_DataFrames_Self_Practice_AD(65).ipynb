```python
import pandas as pd
```

```python
row_labels=[0,1,2,3,4]
column_labels=['A', 'B', 'C']

#actually looking at above labels, it can be inferred, that data will have 5rows, and 3 column, so basiccaly array of 5*3

data=[[1,2,5],[4,2,4],[6,7,9],[3,2,4],[3,3,2]]
```

```python
#lets pull it in df, as a DataFrame
```

```python
df=pd.DataFrame(data=data, index=row_labels, columns=column_labels)
```

```python
#lets check the type

type(df)
```

```python
df
```

```python
#set index fuction
#pandas.set_index(keys, drop=true_by_default, append=False_by default, inplace=False_by default, ..other..)

#key=column or array of the same length, that shd be made an index now
#drop=this drops the key column,as key is now moved to index.Make drop=false,if you need the column (as key and column both)
#append=if you need column to be append to existing key, set is as true, otherwise, it is false, bydefault
#implace=inplace=true, changes the df value, if set as false(by default), original df remains intact

#lets see, lets set column 'C' as index
```

```python
df.set_index('C') #even keys='C' would have worked, but the position is same, so only 'C' will work too
```

```python
#cool, 'C' is set as index, but if we also need 'C' as column, use:

df.set_index('C', drop=False)
```

```python
#now if we call df again, you will see changes are NOT made there

df
```

```python
#to make the changes in existing dataframe (df) use:

df.set_index(key='C', drop=False, inplace=True)
```

```python
#now, if you see, df changed :)


df
```

```python
#to append key'c' with 'B', use: append=true (with key='B' now)

df.set_index('B', drop=False, append=True)
```

```python
#cool, b is now appended to C
```

```python
#reset-index ( to reset the index) use:

#df.reset_index(level=None, drop=False, inplace=False, ..other..)

#level=none, removes all level by default
#drop=False reinserts index. If true, iit drops the index and won't reinsert

#lets see below:
```

```python
df
```

```python
#lets reset index in df

df.reset_index() #expecting error (as column C already exists, so we may have to drop it, but lets see..)
```

```python
#error it is, coz column 'C' already exists, and once it try to reinsert index 'C' as column 'C', we get error..
#..so lets dro the index, so make drop=True
```

```python
df.reset_index(drop=True)
```

```python
#cool, it worked...but note this is not inplace now, as 'df' is changed, to keep this value, lets specfy inplace=True
```

```python
df.reset_index(drop=True, inplace=True)
```

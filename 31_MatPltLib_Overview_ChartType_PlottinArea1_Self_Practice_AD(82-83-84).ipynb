```python
# twinx method (one axes with 2-yaxis scenerio).

# lets say if we want to merge above chart 1 & chart 2, but keep seperate axis, we use twin() for this
```

```python
fig=plt.figure(figsize=(4,4))

axes1=fig.add_axes(rect=[0,0,1,1])
axes1.set_ylim(0,100)
axes1.set_title('Chart 1')
axes1.set_xlabel('X-1')
axes1.set_ylabel('Y-1')

axes2=axes1.twinx() #..so see here we just created axes2 on top of axes1, instead of creating it on top of fig object

```

```python
# lets define axes2 lower & upper limits..AND new Y-label

fig=plt.figure(figsize=(4,4))

axes1=fig.add_axes(rect=[0,0,1,1])
axes1.set_ylim(0,100)
axes1.set_title('Chart 1')
axes1.set_xlabel('X-1')
axes1.set_ylabel('Y-1')

axes2=axes1.twinx()
axes2.set_ylim(0,50)
axes2.set_ylabel('Y-2')
```

```python
#Cool, lets move to (to and from) HTML 

# -- pd.read_html

# -- read tables from websites
# -- stores in as list, then can be called as list[0] or list[1] or better as list[0] if matching table is called
```

```python
#lets dig into wikipedia page for - List_of_highest-grossing_films
```

```python
url='https://en.wikipedia.org/wiki/List_of_highest-grossing_films'
```

```python
movies=pd.read_html(url)
```

```python
type(movies)
```

```python
#so this is a list object, which can be indexed to find out table of choice (or use match argument to filter down)
```

```python
movies[0]
```

```python
#cool, so we pulled the first table, but what if we want the table 'High-grossing films by year of release[47][48][49]'
#..for this we will have to use 'match' argument in read_html
```

```python
movies=pd.read_html(url, match='High-grossing films by year of release')
```

```python
#still this will give us the list, but at index 0, we should have our table
```

```python
movies[0]
```

```python
#and it is there at index=0, cool
```

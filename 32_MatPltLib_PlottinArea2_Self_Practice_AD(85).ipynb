```python
fig,ax=plt.subplots(1,2) #just calling to refer, will modify in the next line
```

```python
fig,ax=plt.subplots(1,2) #creating 1*2 array, so access using ax[0],ax[1]

ax[0].set_title('Axes_Chart1')
ax[1].set_title('Axes_Chart2')
```

```python
# cool, lets modify other stuff too

fig,ax=plt.subplots(1,2)

ax[0].set_title('Axes_Chart1')
ax[0].set_xlim(0,10)
ax[0].set_ylim(0,20)
ax[0].set_xlabel('X1')
ax[0].set_ylabel('Y1')



ax[1].set_title('Axes_Chart2')
ax[1].set_xlim(0,30)
ax[1].set_ylim(0,40)
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')
```

```python
#lets make it look little clean with 'tight_layout=True'

fig,ax=plt.subplots(1,2,tight_layout=True)

ax[0].set_title('Ax[0]_Chart1')
ax[0].set_xlim(0,10)
ax[0].set_ylim(0,20)
ax[0].set_xlabel('X1')
ax[0].set_ylabel('Y1')



ax[1].set_title('Ax[1]_Chart2')
ax[1].set_xlim(0,30)
ax[1].set_ylim(0,40)
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')
```

```python
# cool stuff so far

# we can also, instead of ax[0] & ax[1], un pack it like fig, [ax1,ax2], and then independently access ax1&ax2, lets do

# note this applicable only when ax is 2 or more dimensional array
```

```python
fig, [ax1,ax2]=plt.subplots(1,2)
```

```python
fig, [ax1,ax2]=plt.subplots(1,2)

ax1.set_title('Ax1_Chart1')
ax2.set_title('Ax2_Chart2')
```

```python
#cool, similarly we can set its xlim, xlabel, ylims, ylabel etc....lets move on to 2*2 axes & see how they work
```

```python
fig,ax=plt.subplots(2,2, tight_layout=True)
```

```python
#lets assign chart label

fig,ax=plt.subplots(2,2, tight_layout=True) 

# This creates 1*2 array, so access using ax[0,0],ax[0,1],ax[1,0],ax[1,1] like we do for numpy array

ax[0,0].set_title('Ax[0,0]_Chart1')
ax[0,1].set_title('Ax[0,1]_Chart2')
ax[1,0].set_title('Ax[1,0]_Chart3')
ax[1,1].set_title('Ax[1,1]_Chart4')
```

```python
# that is so cool, similary what ever dimension array we get (2*2), 3*3 or 4*4 or so on, we can access the components...

# ...and set labels, lims, title, axis_value accordingly..
```

```python
# other way to achieve above result is to use [ax1,ax2] isntead of just ax, saves some slicing the array thru indexing...
#...so instead of 'ax' component sliced/indexed like ax[0,0], ax[0,1], ax[1,0], ax[1,1]...
#...we can now slice/index it with ax1[0], ax1[1], ax2[0], ax2[1]..little less effort
```

```python
#cool, lets see how to do this where ax creates 2*2 array
```

```python
fig, [ax1,ax2]=plt.subplots(2,2, tight_layout=True)
```

```python
fig, [ax1,ax2]=plt.subplots(2,2, tight_layout=True)

ax1[0].set_title('Ax1[0]_Chart1')
ax1[1].set_title('Ax1[1]_Chart2')
ax2[0].set_title('Ax2[0]_Chart3')
ax2[1].set_title('Ax2[1]_Chart4')
```

```python
#lets try unpacking 2*2 further, using [[ax1,ax2],[ax3,ax4]]...probably we wond have to slice/index then..lets see

# other way to achieve above result is to use [ax1,ax2] isntead of just ax, saves some slicing the array thru indexing...
#...so instead of 'ax' component sliced/indexed like ax[0,0], ax[0,1], ax[1,0], ax[1,1]...
#...we can now slice/index it with ax1[0], ax1[1], ax2[0], ax2[1]..little less effort
```

```python
#cool, it worked better, no slicing/indexing the array to set chart parameters/attributes:

#but, lets fix the aboce chart for better visual (tight_layout=True)
```

```python
#cool, played along enough on concatenate, lets move to join

#Joining DataFrame (specially in SQL)

# DataFrame.join(other, on, how, ..OtherOptionalParameter..)
# other = the Other DataFrame, the join key must be an index
# on = key to join on
# how = 'inner', 'outer', 'left', 'right'...default=left
```

```python
emp=pd.read_excel('employees_hr.xls', sheet_name='employees') #installed pip install xlrd, then it worked
```

```python
dept = pd.read_excel(io='employees_hr.xls', sheet_name='departments')
```

```python
dept
```

```python
emp.head()
```

```python
dept.head()
```

```python
emp.join(dept, on='dept_id', how='left') #it worked, but this is NOT the way to join
```

```python
# as mentioned... other = the Other DataFrame, the join key must be an index
# this means, that in dept DataFrame,id must be set as index, coz 1st df will always join on index of 2nd df
# it worked above when i tried, as original index was same as id field

#so , we have to set index of other df (i.e. dept df)
```

```python
emp.join(dept.set_index('id'), on='dept_id', how='left')
```

```python
#lets say we dont want all columns to be returned, just limited columns, then just stack those (columns) up..
#..like list of list [['a', 'b','c']]
```

```python
emp.join(dept.set_index('id'), on='dept_id', how='left')[['emp_id', 'dept_name']]
```

```python
# now, merge functions
# same as join, advantage over join is that we dont need to reset index (cooool!!)
# pd.merge(left, right, left_on, right_on, how, ..other_optional..)
# left= left df; right=right df; left_on=key to join on, from left df; right_on=key to join on, from right df...
#...how= 'left'/'right'/'outer' whatever type of join needed (left=by default)
```

```python
emp.head(2)
```

```python
dept.head(2)
```

```python
pd.merge(left=emp, right=dept, left_on='dept_id', right_on='id', how='left')
```

```python
pd.merge(left=emp, right=dept, left_on='dept_id', right_on='id', how='right')
```

```python
#now, lets say if we only want ['emp_id', 'dept_name', 'dept_location'] from the table, we will do it ike this..
```

```python
pd.merge(left=emp, right=dept, left_on='dept_id', right_on='id', how='right')[['emp_id', 'dept_name', 'dept_location']]
```

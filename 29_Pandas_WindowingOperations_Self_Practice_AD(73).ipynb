```python
#now, lets sort the 'Month Year' column
```

```python
tfl_df.sort_values('Month Year', inplace=True)
```

```python
tfl_df
```

```python
#cool, now lets group by 'Month Year' (sum)
```

```python
tfl_12_month = tfl_df.groupby(by='Month Year').sum()
```

```python
tfl_12_month
```

```python
tfl_12_month.tail(12) # lookin at last 12 rows,JLT
```

```python
#resetting index for last 12 months (basically in above DataFrame, and updating it with it)
```

```python
tfl_12_month = tfl_12_month.tail(12).reset_index()
```

```python
tfl_12_month
```

```python
# cool, now windowing operations

# windowing Operations (expanding - for cumulative sum)
# df.expanding(..optional arguments..) ---> this suggest cumulative operation down the row
# df.expanding(..optional arguments..).sum() ---> this suggest cumulative 'sum' operation down the row (similarly, do mean, max, min, std.. etc)
```

```python
tfl_12_month['Number of Bicycle Hires'].expanding().sum()
```

```python
#cool, lets add this as a new column to existing DataFrame (tfl_12_months['cumulative'])
```

```python
tfl_12_month['cumulative'] = tfl_12_month['Number of Bicycle Hires'].expanding().sum()
```

```python
tfl_12_month
```

```python
# rolling operation
# similar to cumulative, but is done for rolling 3 months or rolling 6 months or so
# basicaaly cumulates for 3 rows or 6 rows intervals (instead of cumulating all the way till last row)
```

```python
#cool, lets do it in above DataFrame (create a rolling 3 months and 6 months total)
```

```python
tfl_12_month['Rolling Sum 3']=tfl_12_month['Number of Bicycle Hires'].rolling(3).sum()
```

```python
tfl_12_month['Rolling Sum 6']=tfl_12_month['Number of Bicycle Hires'].rolling(6).sum()
```

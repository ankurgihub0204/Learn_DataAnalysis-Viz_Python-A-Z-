```python
import pandas as pd
```

```python
countries_df=pd.read_csv('top_10_countries_2.csv')
```

```python
countries_df.dtypes
```

```python
#cool, population is integer here, grt!
```

```python
#dropping '% of world' & 'Date' column from DataFrame
```

```python
countries_df.drop(columns=['% of world', 'Date'], inplace = True)
```

```python
countries_df
```

```python
#cool, columns are dropped
```

```python
#Renaming 'Country / Dependency' to 'Country'
```

```python
countries_df.rename(columns={'Country / Dependency':'Country'}, inplace =True)
```

```python
countries_df
```

```python
#cool, columns renamed
```

```python
#now, insert new column for 'Sub Region'
```

```python
sub_regions = ['Eastern Asia', 'Southern Asia', 'Northern America',
               'Southeast Asia', 'Southern Asia','Southern America',
               'Western Africa','Southern Asia','Eastern Europe','Central America']
```

```python
countries_df['Sub Region']=sub_regions
```

```python
countries_df
```

```python
#changing the order of columns (this time using reindex, not loc)
```

```python
countries_df.columns
```

```python
countries_df = countries_df.reindex(columns=['Rank', 'Country', 'Region', 'Sub Region', 'Population'])
```

```python
countries_df
```

```python
countries_df.dtypes
```

```python
#changing 'Country', 'Region' & 'Sub Region' to string type
```

```python
countries_df[['Country', 'Region']]=countries_df[['Country', 'Region']].astype('string')
```

```python
countries_df['Sub Region']=countries_df['Sub Region'].astype('string')
```

```python
countries_df.dtypes
```

```python
# groupby method

# df.groupby(by, ..others..)  by=column or columns to be grouped (pass single or)
```

```python
#lets group by by region
```

```python
countries_df.groupby(by='Region')
```

```python
#oh, this return an object, so lets appy an aggregation to this object, lets apply 'sum'
```

```python
countries_df.groupby(by='Region').sum()
```

```python
#hmm shoot, what all other column are doing here, so we will have to do like this...
```

```python
countries_df[['Region', 'Population']].groupby(by='Region').sum()
```

```python
countries_df
```

```python
#lets do it by region by Sub Region
```

```python
countries_df[['Region', 'Sub Region','Population']].groupby(by=['Region', 'Sub Region']).sum()
```

```python
#similarly, we can group by for (mean, max, std, describe etc)...lets see describe below
```

```python
countries_df[['Region', 'Sub Region','Population']].groupby(by=['Region', 'Sub Region']).describe()
```

```python
#lets say, if we want to group by and do sum, mean, max (all in one go)..then..

#use aggregation method - agg([func]), instead of sum()

# not sure why 'mean' and 'count' needs to be in '', where as other are not :(
```

```python
countries_df[['Region', 'Sub Region','Population']].groupby(by=['Region', 'Sub Region']).agg([sum, 'mean', max, 'count'])
```

```python
#pivoting the data

#pd.pivot_table(data, values, aggfunc, index, columns, fill_value, ...)

# data = DataFrame; # Values = columns to be aggregated; # aggfunc = aggregation function ( by deafult - mean).
# index = columns to be displayed as index (single or multiple columns)
# columns = column to be displayed along columns (single or multiple columns)
# fill_value = if the result is NA, fill it with something (may be fill it with 0)
```

```python
countries_df
```

```python
#lets say now what we want is 'Country' as rows, 'Sub Region' as columns, and the 'Population' (sum) value in the matrix
```

```python
pd.pivot_table(data=countries_df, values='Population', aggfunc=sum, index='Country', columns='Sub Region')
```

```python
#cool, lets say we dont want 'NaN', so lets relace (or fill) it with 0
```

```python
pd.pivot_table(data=countries_df, values='Population', aggfunc=sum, index='Country', columns='Sub Region', fill_value=0)
```

```python
#cool, so lets practice the pivot table

#lets pivot it in a way that it looks like our group by table, as below
```

```python
countries_df[['Region', 'Sub Region','Population']].groupby(by=['Region', 'Sub Region']).sum()
```

```python
pd.pivot_table(data=countries_df, index=['Region', 'Sub Region'], values='Population', aggfunc=sum)
```

```python
# we can also achiev this by pivot_table as a method
```

```python
countries_df.pivot_table(values='Population', index=['Region', 'Sub Region'], aggfunc=sum)
```

```python
players_100=players.head(100)
```

```python
#JLT, plotting scatter plot to see co-relation between players_100  overall rating(x) & wages(y)
```

```python
plt.scatter('overall', 'wage_eur', data=players_100)
```

```python
#other way to do it:

plt.scatter(players_100['overall'], players_100['wage_eur'])
```

```python
#OR...other way to do it:

plt.scatter(x=players_100['overall'], y=players_100['wage_eur'])
```

```python
#cool, now lwts stick to tutorial, and use fig, ax objects to draw same...
```

```python
fig=plt.figure(figsize=(5,5))
ax1=fig.add_axes(rect=[0,0,1,1])
ax1.set_title('Overall ability vs wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('wages')

#plotting scatter plot now:

ax1.scatter(x=players_100['overall'], y=players_100['wage_eur'])
plt.show()
```

```python
#now here, we can size the bubble to correspond to anther dimension too, using argument 's', lets size it per player's value
```

```python
#but first, lets see how size 's' effects the chart, we can assign integer value to it to increase bubble size
```

```python
fig=plt.figure(figsize=(5,5))
ax1=fig.add_axes(rect=[0,0,1,1])
ax1.set_title('Overall ability vs wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('wages')

#plotting scatter plot now:

ax1.scatter(x=players_100['overall'], y=players_100['wage_eur'], s=120)
plt.show()

#so bubble size is bigger now, lets assign s= player's value
```

```python
fig=plt.figure(figsize=(5,5))
ax1=fig.add_axes(rect=[0,0,1,1])
ax1.set_title('Overall ability vs wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('wages')

#plotting scatter plot now:

ax1.scatter(x=players_100['overall'], y=players_100['wage_eur'], s=players_100['value_eur']/1000000, color='r')
plt.show()

#hmm, this is because bubble size is now so- so big, that it took entire canvas(or plotting area), lets divide it by 1-mill
```
